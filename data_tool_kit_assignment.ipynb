2 How does broadcasting work in NumPy?


ans

    Broadcasting in NumPy describes how arrays with different shapes are
    
    treated during arithmetic operations. It allows element-wise operations to
    
    be performed on arrays that do not have the exact same shape, provided
    
    their shapes are "compatible" according to a specific set of rules.


---


**How Broadcasting Works:**
    Dimension Alignment (Right to Left): NumPy compares the shapes of the
    
    arrays starting from the trailing (rightmost) dimension and moving towards
    
    the leading (leftmost) dimension.


**Compatibility Rules:**

    For each dimension being compared, two arrays are considered compatible if:

    The dimension sizes are equal.

    One of the dimension sizes is 1. In this case, the array with the dimension
  
    of size 1 is "stretched" or "broadcast" along that dimension to match the

    size of the other array. This stretching happens virtually, without
    
    actually copying data, making it memory efficient.


**Dimension Padding:**

    If arrays have a different number of dimensions, the array with fewer
    
    dimensions is conceptually padded with ones on its leading (left) side
    
    until it has the same number of dimensions as the larger array.


**Error Handling:**

    If, for any dimension, the sizes are not equal and neither of them is 1,
    
    then the arrays are not compatible for broadcasting, and a ValueError is raised.

4. Explain the use of the groupby() method in Pandas

ans


    The groupby() method in Pandas is a powerful and versatile function used
    
    for splitting a DataFrame into groups based on one or more criteria,
    
    applying a function to each group independently, and then combining the
    
    results into a new DataFrame. This "split-apply-combine" strategy is
    
    fundamental for data analysis and manipulation.


**Core Functionality:**




---


**Splitting**:

    The groupby() method first divides the DataFrame into sub-DataFrames based
    
    on the unique values in the specified column(s). For example, grouping by a
   
    'City' column would create separate groups for each city.

**Applying**:

    After splitting, a function is applied to each individual group. This
    
    function can be an aggregation function (like sum(), mean(), count(), min
    
    (), max()), a transformation function (like transform()), or a filtration
    
    function (like filter()).

**Combining**:

    Finally, the results from applying the function to each group are combined
    
    back into a single Series or DataFrame, often with the group keys as the
    index.


**Common Use Cases:**




```python
#Aggregation:

   # Calculating summary statistics for each group (e.g., average sales per

   # region, total count of items per category).

   # example >>>>>


import pandas as pd
data = {'Category': ['A', 'B', 'A', 'B', 'A'], 'Value': [10, 20, 15, 25, 12]}
df = pd.DataFrame(data)
grouped_sum = df.groupby('Category')['Value'].sum()
print(grouped_sum)
```

T**ransformation:**

    Performing group-specific computations while maintaining the original shape of the data (e.g., standardizing values within each group, filling missing values within groups).

    

```python
#example

import pandas as pd
data = {'Group': ['X', 'Y', 'X', 'Y'], 'Score': [80, 90, 75, 85]}
df = pd.DataFrame(data)
df['Normalized_Score'] = df.groupby('Group')['Score'].transform(lambda x: (x - x.mean()) / x.std())
print(df)
```

**Filtration**:

    Discarding or keeping groups based on a group-wise condition (e.g., keeping
    
    only groups with a minimum number of members, filtering out groups where
    
    the sum of a column is below a certain threshold).

```python
    import pandas as pd
    data = {'ID': [1, 2, 3, 4, 5], 'Group': ['A', 'A', 'B', 'A', 'B'], 'Value': [10, 5, 20, 12, 8]}
    df = pd.DataFrame(data)
    filtered_df = df.groupby('Group').filter(lambda x: len(x) > 2)
    print(filtered_df)
```

6. What are the differences between NumPy arrays and Python lists?

  ans


**Data Type Homogeneity:**

---




**NumPy Arrays:**

    Must contain elements of a single, uniform data type (e.g., all integers,
    
    all floats). This homogeneity enables optimizations for numerical
    
    operations.

**Python Lists:**

    Can contain elements of different data types (e.g., integers, strings,
    
    other lists) within the same list.



---


**Performance and Memory Efficiency**:

---




**NumPy Arrays:**

    Optimized for numerical computations and large datasets. They are more
    
    memory-efficient and significantly faster for mathematical operations due
    
    to their fixed size, homogeneous data, and contiguous memory storage.

**Python Lists:**


    Less efficient for numerical computations, especially with large datasets,
    
    due to the overhead of storing heterogeneous data and potential memory
    
    fragmentation.


**Functionality and Operations:**


---




**NumPy Arrays:**

    Offer a vast array of built-in functions and methods for advanced
      

    mathematical and statistical operations (e.g., linear algebra, Fourier
    
    transforms, broadcasting for element-wise operations).


**Python Lists**:

    Provide general-purpose list manipulation methods but lack specialized
    
    functions for numerical computations. Arithmetic operations on lists
    
    typically involve explicit looping or list comprehensions.


**Flexibility**:



---


**NumPy Arrays:**

    Less flexible in terms of data type and size modification once created, as
    
    they are fixed-type and fixed-size by nature.


**Python Lists:**


    Highly flexible, allowing dynamic resizing and modification of elements
    
    with different data types.


**When to Use Which:**


**Use NumPy Arrays:**

    For numerical computing, scientific applications, and working with large
    
    datasets where performance and memory efficiency are crucial.


**Use Python Lists**:

    For general-purpose data storage, when data heterogeneity is required, or
    
    when the dataset size is relatively small and performance is not a primary
    
    concern.

15. How does NumPy handle multidimensional arrays?

ans


**Here's how NumPy handles multidimensional arrays:**


**Creation:**

    Multidimensional arrays are typically created by passing nested Python
    
    lists to np.array(). For example, a list of lists creates a 2D array
    
    (matrix), and a list of lists of lists creates a 3D array. Functions like
    
    np.zeros(), np.ones(), or np.arange() can also create arrays of specified
    
    shapes, which can then be reshaped using reshape().
    



```python
#example

import numpy as np

# Create a 2D array
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix)

# Create a 3D array of zeros
three_d_array = np.zeros((2, 3, 4))
print(three_d_array)
```

**Internal Storage:**

    While conceptually multidimensional, the data within a NumPy array is
    
    actually stored in a flat, one-dimensional block of memory. The shape and
    
    strides attributes provide the metadata needed to interpret this flat
    
    memory as a multidimensional structure.


**Indexing and Slicing:**

    Elements in multidimensional arrays are accessed using multi-dimensional
    
    indexing with a single set of square brackets and comma-separated indices
    
    for each dimension. Slicing works similarly, using colons to specify ranges
    
    within each dimension.

```python
#example

import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])

# Access an element
print(arr[0, 1]) # Output: 2

# Slice a sub-array
print(arr[:, 1:3]) # Output: [[2, 3], [5, 6]]
```

**Operations:**

    Mathematical and logical operations on multidimensional arrays are
    
    performed element-wise by default. NumPy's broadcasting rules allow
    
    operations between arrays of different but compatible shapes, automatically
    
    replicating values along necessary dimensions.


**Reshaping and Views:**

    NumPy allows changing the shape of an array without copying the underlying
    
    data, as long as the total number of elements remains the same. This
    
    creates a "view" of the original data with a different interpretation of
    
    its dimensions.

```python
#example

import numpy as np

arr = np.arange(6) # [0, 1, 2, 3, 4, 5]
reshaped_arr = arr.reshape((2, 3))
print(reshaped_arr)
```

17. A Explain the difference between apply() and map() in Pandas?

ans

In Pandas, both map() and apply() are used to transform data, but they operate on different data structures and have slightly different functionalities.


**map() (Series only):**
**Purpose:**


    Primarily used for element-wise substitution of values within a Pandas
    
    Series. It's ideal for mapping existing values to new ones, often using a
    
    dictionary, another Series, or a function.


**Operation:**

    It iterates through each element of a Series and replaces it based on the
    
    mapping provided. If a value in the Series is not found in the mapping, it
    
    will be replaced with NaN by default.

```python
#example
import pandas as pd
s = pd.Series([1, 2, 3, 4])
mapping = {1: 'A', 2: 'B'}
result = s.map(mapping)
print(result)
# Output:
# 0      A
# 1      B
# 2    NaN
# 3    NaN
# dtype: object
```

**apply() (Series and DataFrame):**


**Purpose**:

 A more general-purpose method for applying a function along an axis of a

 DataFrame (row-wise or column-wise) or to the values of a Series. It's

 suitable for more complex operations that might not be simple one-to-one mapping.
**Operation:**


**On a Series:**

 It applies a function to each individual element of the Series, similar to map() but offers more flexibility in the function's complexity.


**On a DataFrame:** It applies a function to each row or each column of the DataFrame. The function receives either a Series (for each row/column) or a single value (if applied element-wise to a specific column).

```python
#example on a serise

import pandas as pd
s = pd.Series([1, 2, 3, 4])
def square(x):
    return x * x
result = s.apply(square)
print(result)
# Output:
# 0     1
# 1     4
# 2     9
# 3    16
# dtype: int64
```

```python
#example (on a data frame)

import pandas as pd
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
def sum_row(row):
    return row['A'] + row['B']
result = df.apply(sum_row, axis=1) # axis=1 for row-wise application
print(result)
# Output:
# 0    4
# 1    6
# dtype: int64
```

**Key Difference**


**Scope**:

    map() is exclusively for Series, while apply() works on both Series and
    
    DataFrames.


**Functionality:**  

    map() is optimized for simple value substitution or mapping, whereas apply
    
    () provides more flexibility for applying arbitrary functions, including
    
    those requiring row/column context in DataFrames.

21. ✈ Why is NumPy’s array slicing faster than Python’s list slicing?


ans


**NumPy's array slicing is significantly faster than Python's list slicing due to several key architectural differences:**



**Contiguous Memory Allocation:**

    NumPy arrays store elements of the same data type in a contiguous block of
    
    memory. This "locality of reference" allows for efficient access and
    
    manipulation, as the processor can quickly retrieve data without "chasing
    
    pointers" to different memory locations. Python lists, on the other hand,
    
    store pointers to objects that can be scattered throughout memory,
    
    requiring more overhead to access elements.


**Homogeneous Data Types:**

    NumPy arrays enforce a single data type for all elements within an array.
    
    This eliminates the need for type checking during operations and allows
     
    NumPy to optimize memory usage and computations at a lower level (often
     
    implemented in C or Fortran). Python lists can hold heterogeneous data
    
    types, requiring more general-purpose handling and type checks during
    
    operations.


**Vectorization and Optimized C Code:**


    NumPy leverages vectorization, allowing operations to be applied to entire
    
    arrays simultaneously rather than element by element. These vectorized
     
    operations are implemented in highly optimized C or Fortran code,
     
    providing a substantial performance boost compared to Python's interpreted
    
    loops for list operations.


**Reduced Overhead**:

    Python lists carry overhead for each element, including type information
    
    and reference counts. NumPy arrays, with their fixed data types and
    
    contiguous storage, minimize this overhead, leading to more
    
    memory-efficient and faster operations, especially with large datasets.

7. Write a program to perform element-wise multiplication of two NumPy arrays?

```python
import numpy as np

# Create two NumPy arrays
array1 = np.array([1, 2, 3, 4])
array2 = np.array([5, 6, 7, 8])

# Method 1: Using the * operator
result_operator = array1 * array2
print("Result using * operator:", result_operator)

# Method 2: Using np.multiply() function
result_function = np.multiply(array1, array2)
print("Result using np.multiply():", result_function)

# Example with 2D arrays
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

result_matrix = matrix1 * matrix2
print("\nResult with 2D arrays (using * operator):\n", result_matrix)
```

11. Perform matrix multiplication using NumPy

```python
import numpy as np

# Define two matrices as NumPy arrays
matrix_a = np.array([[1, 2, 3],
                     [4, 5, 6]])

matrix_b = np.array([[7, 8],
                     [9, 10],
                     [11, 12]])

print("Matrix A:")
print(matrix_a)
print("\nMatrix B:")
print(matrix_b)

# Method 1: Using np.dot()
result_dot = np.dot(matrix_a, matrix_b)
print("\nResult using np.dot():")
print(result_dot)

# Method 2: Using the @ operator (Python 3.5+)
result_at = matrix_a @ matrix_b
print("\nResult using the @ operator:")
print(result_at)
```

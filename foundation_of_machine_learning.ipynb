2- : Explain overfitting and underfitting in ML. How can you detect and prevent
them?

ans-


**Overfitting**:


    Overfitting occurs when a model learns the training data too well,
    
    including noise and specific patterns that are not representative of the
    
    underlying data distribution. This results in excellent performance on the
    
    training set but poor performance on new, unseen data. The model
    
    essentially memorizes the training examples rather than learning
    
    generalizable patterns.


**Underfitting:**


    Underfitting occurs when a model is too simple to capture the underlying
    
    patterns in the data. It fails to learn the relationships between features
    
    and the target variable, leading to poor performance on both the training
    
    and test sets. The model is unable to adequately represent the complexity
    
    of the data.


**Detection -**


 **evaluate on training and validate / tests sets :**


    the primary method for detecting both is by compairing the both models
    
    performance on training sets and a seprate validation on or test set
    

**Overfitting**-

    High accuracy/low error on the training set, but significantly lower
    
    accuracy/higher error on the validation/test set.



**underfitting**-

    Low accuracy/high error on both the training and validation/test sets.


**Prevention**:


**To prevent Overfitting:**


**Regularization**:

    Techniques like L1 (Lasso) or L2 (Ridge) regularization add a penalty to
    
    the model's loss function for large weights, encouraging simpler models.


**More Data**:

    Increasing the size and diversity of the training data can help the model
    
    learn more generalizable patterns.


**Feature Selection/Reduction**:

    Reducing the number of features or selecting the most relevant ones can
    
    prevent the model from learning noise.


**Early Stopping**:


    Monitoring the validation loss during training and stopping when it starts
    
    to increase can prevent the model from overfitting.


Simpler Model Architectures:

    Using models with fewer layers or parameters if the current model is overly
    
    complex for the data.



**Cross-validation**:


    Using techniques like k-fold cross-validation helps assess the model's
    
    
    generalization performance more robustly.



---


**To prevent Underfitting:**



 **Increase Model Complexity**:


    Use a more complex model architecture (e.g., more layers in a neural
    
    network, a higher-degree polynomial for regression).


**Add More Features:**


    Introduce new features that might provide more information to the model
    
    about the underlying relationships in the data.


**Reduce Regularization:**


    If regularization was applied, reducing its strength can allow the model to
    
    learn more complex patterns.


**Train longer**-

    In some cases, simply training the model for more epochs can help it
    
    converge to a better solution, especially for neural networks.






3- :How would you handle missing values in a dataset? Explain at least three
methods with examples.


ans-




    Handling missing values is a crucial step in data preprocessing.
    
    Here are three methods with examples:


**1. Deletion:**

    This involves removing rows (samples) or columns (features) that contain
    
    missing values.


**When to use:**


    Row Deletion:   
    
    When a small percentage of rows have missing values and the missingness is
    
    random, or when the missing values are in the target variable.


    Column Deletion:
    
    
    When a column has a very high percentage of missing values (e.g., >70-80%),
    
    making it unreliable for analysis.

```python
    import pandas as pd
    import numpy as np

    data = {'A': [1, 2, np.nan, 4],
            'B': [5, np.nan, 7, 8],
            'C': [9, 10, 11, np.nan]}
    df = pd.DataFrame(data)

    # Delete rows with any missing values
    df_rows_dropped = df.dropna()
    print("DataFrame after dropping rows:\n", df_rows_dropped)

    # Delete columns with any missing values
    df_cols_dropped = df.dropna(axis=1)
    print("\nDataFrame after dropping columns:\n", df_cols_dropped)
```

2. **Imputation with Central Tendency (Mean/Median/Mode):**


**Method**:


    Replace missing values with the mean, median (for numerical data), or mode
    
    (for categorical data) of the respective feature.


**When to use**:

    When the missingness is assumed to be random and the distribution of the
    
    feature is not heavily skewed (for mean imputation) or when a quick and
    
    simple imputation is needed.

```python
    import pandas as pd
    import numpy as np

    data = {'Age': [25, 30, np.nan, 40, 28],
            'City': ['NY', 'LA', 'NY', np.nan, 'SF']}
    df = pd.DataFrame(data)

    # Impute missing 'Age' with the mean
    mean_age = df['Age'].mean()
    df['Age'].fillna(mean_age, inplace=True)

    # Impute missing 'City' with the mode
    mode_city = df['City'].mode()[0] # .mode() can return multiple modes
    df['City'].fillna(mode_city, inplace=True)
    print("DataFrame after mean/mode imputation:\n", df)
```

3. **Advanced Imputation (e.g., K-Nearest Neighbors - KNN Imputation)**:


**Method:**


    Impute missing values based on the values of similar data points
    
    (neighbors). KNN imputation finds the 'k' nearest neighbors to a data point
    
    with a missing value and uses their values to estimate the missing one (e.
    
    g., by averaging for numerical data or taking the most frequent for
    
    categorical).


**When to use:**


    When the missingness is not completely random and there are relationships
    
    
    between features that can be leveraged for more accurate imputation.

    

```python
    import pandas as pd
    import numpy as np
    from sklearn.impute import KNNImputer

    data = {'Feature1': [10, 12, np.nan, 15, 11],
            'Feature2': [20, np.nan, 25, 22, 21]}
    df = pd.DataFrame(data)

    imputer = KNNImputer(n_neighbors=2) # Use 2 nearest neighbors
    df_imputed_array = imputer.fit_transform(df)
    df_imputed = pd.DataFrame(df_imputed_array, columns=df.columns)
    print("DataFrame after KNN imputation:\n", df_imputed)
```

4- What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical)?

ans-



    Imbalanced data pertains to datasets where the distribution of observations
    
    in the target class is uneven. In other words, one class label has a
    
    significantly higher number of observations, while the other has a notably
    
    lower count.

**ways to handle data for classification;**


1**. Different Evaluation Metric**


    Classifier accuracy is calculated by dividing the total correct predictions
    
    by the overall predictions, suitable for balanced classes but less
     
    effective for imbalanced datasets. Precision gauges the accuracy of a
     
    classifier in predicting a specific class, while recall assesses its
    
    ability to correctly identify a class. In imbalanced datasets, the F1 score
    
    emerges as a preferred metric, striking a balance between precision and
    
    recall, providing a more comprehensive evaluation of a classifier's
    
    performance. It can be expressed as the mean of recall and accuracy.

 **2. Resampling (Undersampling and Oversampling)**


    This method involves adjusting the balance between minority and majority
    
    classes through upsampling or downsampling. In the case of an imbalanced
    
    dataset, oversampling the minority class with replacement, termed
    
    oversampling, is employed. Conversely, undersampling entails randomly
    
    removing rows from the majority class to align with the minority class.

    

```python
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import pandas as pd

# Step 1: Create an imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2,
                           weights=[0.9, 0.1], flip_y=0, random_state=42)

# Step 2: Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Define resampling strategy
over = SMOTE(sampling_strategy=0.2)  # Oversample minority class to 20%
under = RandomUnderSampler(sampling_strategy=0.8)  # Undersample majority class to 80%

# Step 4: Create pipeline
steps = [('over', over), ('under', under), ('model', RandomForestClassifier())]
pipeline = Pipeline(steps=steps)

# Step 5: Train and evaluate
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

print(classification_report(y_test, y_pred))
```

5: Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.
ans -


    Feature scaling is important in machine learning to prevent features with
    
    large values from dominating the model and to improve the performance of
    
    algorithms that rely on distance calculations, such as gradient descent,
    
    k-nearest neighbors, and support vector machines.

 Formula:

    Min-Max Scaling uses the minimum and maximum values of the feature in its formula,
    
    while Standardization uses the mean and standard deviation.
Range:

    Min-Max Scaling restricts feature values to a fixed, bounded range (e.g., [0, 1]),
    
    whereas Standardization does not have a fixed range but centers the data
    
    around 0 with a standard deviation of 1.


Outlier Sensitivity:

    Min-Max Scaling is highly sensitive to outliers because they directly
    
    influence the maximum and minimum values. In contrast, Standardization is
    
    less affected by outliers, which is why it is often preferred when
     
    outliers are present.


Best Use Cases:

    Min-Max Scaling is best for algorithms that need data in a bounded range,

    like neural networks, or when the data is not normally distributed and has

    few outliers. Standardization is better suited for algorithms that assume
    
    a Gaussian (normal) distribution, such as linear regression and PCA.


    

6- Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?

ans-


    Prefer Label Encoding for ordinal data (where categories have a natural
    
    order) or when dealing with high-cardinality features (many unique
    
    categories) to avoid high dimensionality. Prefer One-Hot Encoding for
    
    nominal data (no inherent order) with a manageable number of categories to
    
    prevent the model from assuming false relationships between categories.



**Label Encoding**
**How it works:**


    Assigns a unique integer to each category (e.g., "Low": 0, "Medium": 1, "High": 2).


**When to use:**


    Ordinal data: When there is a natural order to the categories,
    
    such as "low," "medium," and "high," this method accurately reflects the ranking.


    High-cardinality features:
    
    
    When a feature has a very large number of unique categories (like zip
    
    codes), one-hot encoding would create too many new columns. Label encoding
    
    is more memory-efficient in this case.


**When to avoid:**


Nominal data:

    Using label encoding on nominal data (e.g., colors, countries) can mislead
    
    the model into believing there is a numerical relationship or hierarchy
    
    between the categories, which is not true.


---




**One-Hot Encoding**


How it works:

    Creates a new binary (0 or 1) column for each unique category in the
    
    original feature. A '1' indicates the presence of that category for a given
    
    row, and a '0' indicates its absence.


When to use:


Nominal data:


    This is the preferred method for nominal data because it treats each
    
    category as an independent entity and avoids introducing false ordinal relationships.


Low to medium-cardinality features:


    It works well when the number of unique categories is small to moderate,
    
    as it doesn't excessively increase the dimensionality of the dataset.


When to avoid:


    High-cardinality features:
    
    It is not practical for features with many unique categories because it can
    
    lead to a massive increase in the number of columns (high dimensionality),
    
    potentially consuming excessive memory and slowing down the model.

 1: What is the difference between descriptive statistics and inferential
statistics? Explain with examples.


ans-
**Descsriptive Statistic**



    Descriptive statistics refers to the process of summarizing and analyzing data to describe its main features in a clear and meaningful way.
    
    It is used to present raw data in a form that makes it easier to understand and interpret.
    
    Descriptive statistics involves both graphical representations (such as charts and plots) and numerical measures to summarize data effectively.
  
    Unlike inferential statistics, which makes predictions about a population based on a sample, Descriptive statistics is applied to data that is already known.

**Inferential Statistics**


    Inferential statistics involves using data from a sample to make predictions, generalizations, or conclusions about a larger population.
    
    Unlike descriptive statistics, which simply summarizes known data, inferential statistics makes inferences or draws conclusions that go beyond the available data.
    
    It uses probability theory to estimate population parameters and test hypotheses.
    
    By working with a sample, inferential statistics allows researchers to make informed decisions without having to gather data from an entire population.
    

 2: What is sampling in statistics? Explain the differences between random
and stratified sampling.

ans

    Sampling is the statistical process of selecting a subset of a population

    to gather data and make inferences about the entire population, rather
     
    than analyzing the whole group. The primary difference between random
    
    sampling and stratified sampling is that random sampling gives every member
    
    an equal chance of selection without ensuring subgroup representation,
    
    whereas stratified sampling divides the population into relevant subgroups
    
    (strata) and then performs random sampling within each group to guarantee
    
    the representation of all subgroups.


**Sampling in Statistics **



**Purpose:**

    To collect data and make informed decisions about a larger population when
    
    it is too costly, time-consuming, or impractical to study every individual.


**Method:**

    A representative subset of the population is chosen to analyze.


**Random Sampling**
**Definition**:

    A technique where each member of the population has an equal and non-zero
    
    chance of being selected for the sample.


**Goal: **

    To create an unbiased sample that can represent the entire population.


**Limitation:**

    While providing randomness, it does not guarantee the representation of all
    
    subgroups within the population; a specific subgroup might be
    
    overrepresented or entirely missed.



---


**Stratified Sampling**
**Definition:**

    A probability sampling method where the population is first divided into
    
    mutually exclusive subgroups, or strata, based on shared characteristics.


**Process**

    After forming these strata, a random sample is selected from each subgroup.


**Goal:**

    To ensure that every subgroup is represented in the final sample, providing
    
    a more balanced and accurate reflection of the population's structure.


**Example**:

    A study on student opinions might divide the school population by grade
    
    level (freshmen, sophomores, juniors, seniors) and then randomly select an
    
    equal number of students from each grade level.


3: Define mean, median, and mode. Explain why these measures of central
tendency are important.

ans



    The mean, median, and mode are measures of central tendency, which are
    
    single values that represent the center or typical value of a dataset.
    
    They are important statistical tools because they provide different
    
    insights into the data, summarize large datasets, and help in making
    
    comparisons.





---
**definations** >>>>

**MEAN**

    The mean is the arithmetic average of a dataset.
    
    
**How to find it**:

    Sum all the values in the dataset and divide the total by the number of values.
    
**Example:**
  
    For the dataset {2, 3, 5, 5, 8, 10}, the sum is 33. The count is 6.\(\text
    
    {Mean}=\frac{2+3+5+5+8+10}{6}=\frac{33}{6}=5.5\)



---



**MEDIAN**


    The median is the middle value of a dataset when the values are arranged in
    
    numerical order. It is the point that separates the higher half of the data
    
    from the lower half.


**How to find it:**


    For an odd number of values: Arrange the numbers in order and find the
    
    single middle value.


**For an even number of values:**

    Arrange the numbers in order and average the two middle values.


**Example (odd dataset):**

    In the ordered dataset {4, 6, 10, 19, 21}, the median is 10.



**Example (even dataset)**:

    In the ordered dataset {3, 8, 9, 12, 12, 15}, the median is the average of
    
    9 and 12, which is 10.5.


---






**MODE**


    The mode is the value that appears most frequently in a dataset.
    
    A dataset can have one mode, more than one mode, or no mode at all.


**How to find it**:

    Tally the frequency of each value in the dataset and identify the value(s)
    
    with the highest count.


**Example:**

    In the dataset {3, 5, 5, 7, 8, 8, 8, 9}, the mode is 8, as it appears most
    
    often.


**Example (no mode)**:

    In the dataset {1, 2, 3, 4, 5}, there is no mode because no number repeats.


*IMPORTANCE OF CENTRAL TENDANCY MEASURE:*



    These measures are important because they offer different insights into a
    
    dataset, depending on the data's characteristics.



4: Explain skewness and kurtosis. What does a positive skew imply about the data?


ans


    Skewness measures a distribution's lack of symmetry, indicating if it leans
    
    to the left or right, while kurtosis measures the distribution's peakedness
    
    and tail weight. A positive skew signifies that the data has a long, heavy
    
    tail on the right side, with a majority of data points clustered on the
    
    left, leading to the mean being greater than the median and mode.



**Skewness**

---




**What it measures**:

    The asymmetry of a distribution.


**Types:**

    Positive (Right) Skew: The right tail is longer or heavier.


    Negative (Left) Skew: The left tail is longer or heavier.



    Zero Skew: A perfectly symmetrical distribution.


---



**Kurtosis**


**What it measures**:

    The "tailedness" or "peakedness" of a distribution compared to a normal distribution.

     
**Types:**


    Mesokurtic: A normal distribution has a kurtosis value of 3.

    Leptokurtic: A distribution with a more peaked center and heavier tails
    
    than a normal distribution, with a kurtosis greater than 3.


    Platykurtic: A flatter distribution with lighter tails than a normal
    
    distribution, with a kurtosis less than 3.


**What Positive Skew Implies**

 **Longer Right Tail:**

    The distribution has a long tail extending to the right, meaning there are
    
    more extreme values on the higher end of the data.


**Mean > Median**:

    The mean is pulled to the right by these extreme values, making it greater than the median.


**Data Concentration:**

    Most of the data points are concentrated on the left side of the distribution.

6: Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60



```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Calculate covariance
covariance = np.cov(list_x, list_y)[0, 1]

# Calculate correlation coefficient
correlation_coefficient = np.corrcoef(list_x, list_y)[0, 1]

print(f"Covariance: {covariance}")
print(f"Correlation Coefficient: {correlation_coefficient}")
```

 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

```python
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.title('Boxplot of Data')
plt.ylabel('Value')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")
print(f"Identified Outliers: {outliers}")
```

 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and daily sales.


● Explain how you would use covariance and correlation to explore this

relationship.

● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]





ans


    Covariance and correlation show the relationship between advertising spend
    
    and daily sales, with covariance indicating direction (positive if both
     
    increase, negative if one increases and the other decreases) and
     
    correlation providing a standardized measure of both direction and
     
    strength on a scale from -1 (strong negative) to 1 (strong positive).
     
    For the given data, the Python code using the scipy.stats.pearsonr
     
    function will compute the correlation coefficient, confirming the strong
   
    positive relationship.

```python
from scipy.stats import pearsonr

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute the Pearson correlation coefficient and the p-value
# The pearsonr function returns the correlation coefficient and the p-value
correlation_coefficient, p_value = pearsonr(advertising_spend, daily_sales)

# Print the results
print(f"Correlation Coefficient: {correlation_coefficient:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpret the results
if correlation_coefficient > 0:
    print("Interpretation: There is a positive correlation, suggesting that as advertising spend increases, daily sales also tend to increase.")
else:
    print("Interpretation: There is a negative or no significant correlation.")
```

9: Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]


ans


    To understand the distribution of 1-10 scale survey data, use the mean,
    
    median, and standard deviation to measure central tendency and variability,
    
    along with histograms and box plots for visualization. A histogram shows
    
    frequency distribution, while a box plot illustrates quartiles and
    
    potential outliers, giving a comprehensive view of customer satisfaction
    
    levels before a new product launch.


**Summary Statistics**


**Mean**:

    Calculates the average score, providing a general sense of overall
    
    satisfaction.


**Median:**

    The middle score when the data is ordered, which can be more representative
    
    than the mean if there are extreme outliers.


**Standard Deviation**:

    Measures the spread or variability of scores around the mean.
    
    A low standard deviation indicates scores are close to the average,
    
    while a high one means scores are more spread out.


**Range:**

    The difference between the highest and lowest scores, giving a simple
    
    measure of the data's total spread.


**Visualizations**


**Histogram:**


    A bar chart where each bar represents a range of scores (e.g., 1-2, 3-4),
    
    showing the frequency of scores within each range. This helps identify
    
    common satisfaction levels and the overall shape of the distribution.


**Box Plot:**

    A chart that visualizes the five-number summary (minimum, first quartile,


    median, third quartile, maximum), clearly showing the spread, central
    
    tendency, and identifying potential outliers.

```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create the histogram
plt.figure(figsize=(8, 6)) # Set the figure size for better readability
plt.hist(survey_scores, bins=range(1, 12), align='left', rwidth=0.8, color='skyblue', edgecolor='black') # Define bins from 1 to 11 to cover all scores

# Add titles and labels
plt.title('Distribution of Customer Satisfaction Scores')
plt.xlabel('Score (1-10)')
plt.ylabel('Frequency')
plt.xticks(range(1, 11)) # Ensure all scores from 1 to 10 are shown as ticks
plt.grid(axis='y', alpha=0.75) # Add a light grid for easier reading

# Display the plot
plt.show()
```
